```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import glob
import pickle
from astropy.visualization import simple_norm
```

```python
for mod in 'A':
    for file in sorted(glob.glob('epoch*pkl')):
        print(file, mod)
        pickle_in = open(file,"rb")
        dtab = pickle.load(pickle_in)      
        fig, axes = plt.subplots(nrows=2, ncols =2, figsize=[8, 8], sharex = 'all', sharey = 'all')
        
        for det_id in range(4):
            det_key = 'det{}'.format(det_id)
            scaled_im = np.divide(dtab[mod][det_key]['det_im'],dtab[mod][det_key]['exp'])
#            print(scaled_im.min(), scaled_im.max())
            



            if det_id == 0:
                norm = simple_norm(scaled_im, 'sqrt')
                im = axes[0, 1].imshow(np.transpose(scaled_im), norm = norm, origin = 'lower')


                axes[0, 1].set_title(det_key)
                axes[0, 1].set_xlabel('RAWY')
                axes[0, 1].set_ylabel('RAWX')
            if det_id == 1:
                im = axes[0, 0].imshow(np.rot90(np.transpose(scaled_im), 3), norm = norm, origin = 'lower')
#                axes[0, 0].set_title(det_key)
#                 axes[0, 0].set_ylabel('RAWX')
#                 axes[0, 0].set_xlabel('RAWY')
            if det_id == 2:
                im = axes[1, 0].imshow(np.rot90(np.transpose(scaled_im), 2), norm = norm, origin = 'lower')
# #                axes[1, 0].set_title(det_key)
# #                 axes[1, 0].set_ylabel('RAWX')
# #                 axes[1, 0].set_xlabel('RAWY')

            if det_id == 3:
                im = axes[1, 1].imshow(np.rot90(np.transpose(scaled_im), 1), norm = norm, origin = 'lower')
# #                axes[1, 1].set_title(det_key)
# #                 axes[1, 1].set_ylabel('RAWX')
# #                 axes[1, 1].set_xlabel('RAWY')


  



#        fig.colorbar(im, ax=axes.ravel().tolist())
        plt.tight_layout()
        plt.show()
        
```

```python
for mod in 'A':
    for file in sorted(glob.glob('epoch*pkl')):
        print(file, mod)
        pickle_in = open(file,"rb")
        dtab = pickle.load(pickle_in)      
        fig, axes = plt.subplots(nrows=1, ncols =4, figsize=[16, 4])
        
        for det_id in range(4):
            det_key = 'det{}'.format(det_id)
            scaled_im = np.divide(dtab[mod][det_key]['det_im'],1.0)
#            print(scaled_im.min(), scaled_im.max())
            if det_id == 0:
                norm = simple_norm(scaled_im, 'sqrt')
            im = axes[det_id].imshow(scaled_im, norm=norm)
        
            axes[det_id].set_title(det_key)

            if det_id == 0:
                axes[det_id].set_xlabel('RAWX')
                axes[det_id].set_ylabel('RAWY')

        plt.show()
        
        
```
