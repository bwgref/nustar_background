```python
from matplotlib import pyplot as plt
import matplotlib

%matplotlib inline
import astropy.units as u

gain_met = 159663600. # from CALDB file
launch_met=77241600. # 2012-06-13T00:00:00

last_cal_year = ( (gain_met - launch_met)*u.s).to(u.yr)
print(last_cal_year)

font = {'family' : 'normal',
        'size'   : 16}
matplotlib.rc('font', **font)
```

```python
em=1
plt.figure(figsize=(20,10))


for mod in ['fpma', 'FPMB']:
    infile = '{}_det0_line{}.txt'.format(mod, em)
    f = open(infile)
    print(infile)
    fits = []
    errorr = []
    hierr = []
    loerr = []
    for line in f:
    #    print(line)
        fields = line.split()
        fits.append(float(fields[0]))
        hierr.append(float(fields[2]) - float(fields[0]))
        loerr.append(float(fields[0]) - float(fields[1]))    
    f.close()

    plt.xlabel('Years since Launch')
    plt.ylabel('Line Centroid (keV)')
    #plt.plot(range(len(fits)), fits, 'o')
    plt.errorbar(range(len(fits)), fits, loerr, fmt='.', label=mod)

    
plt.axvline(x=last_cal_year.value, color = 'r', linestyle='--')

plt.legend()
plt.title('Det 0')

plt.show()
```

```python
em=1
plt.figure(figsize=(20,10))


for det in range(4):
    for mod in ['fpma', 'fpmb']:
        infile = '{}_det{}_line{}.txt'.format(mod, det, em)
        f = open(infile)
        fits = []
        errorr = []
        hierr = []
        loerr = []
        for line in f:
        #    print(line)
            fields = line.split()
            fits.append(float(fields[0]))
            hierr.append(float(fields[2]) - float(fields[0]))
            loerr.append(float(fields[0]) - float(fields[1]))    
        f.close()

        plt.xlabel('Years since Launch')
        plt.ylabel('Line Centroid (keV)')
        #plt.plot(range(len(fits)), fits, 'o')
        plt.errorbar(range(len(fits)), fits, loerr, fmt='.', label='{}'.format(det))
        break
    plt.legend()

plt.axvline(x=last_cal_year.value, color = 'r', linestyle='--')


plt.title('FPMA')
plt.show()
```
