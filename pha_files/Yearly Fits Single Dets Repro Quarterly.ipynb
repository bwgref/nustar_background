# NOTE:

### This notebook is based on the events from all of the "02" event files

### The data were re-calibrated using the "v05" GAIN files.

### However, they also use the v04 CLC files (the current ones in the CALDB).

### This is the same base as was used to produce the v07 gain files in July 2016
- These only allowed the SLOPE to vary and kept the OFFSET fixed at 0.


- The purpose here is to see if we can discern any gain trends over time.


```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np

%matplotlib inline
import astropy.units as u

gain_met = 159663600. # from CALDB file
launch_met=77241600. # 2012-06-13T00:00:00

last_cal_year = ( (gain_met - launch_met)*u.s).to(u.yr)
print(last_cal_year)

font = {'family' : 'normal',
        'size'   : 16}
matplotlib.rc('font', **font)
```

```python
em=3
plt.figure(figsize=(20,10))
# This cell plots the trends for all four detectors vs time for a line set by the "em" variable above.

for det in range(4):
    for mod in ['fpma', 'fpmb']:
        infile = '{}_det{}_line{}_repro_epochs.txt'.format(mod, det, em)
        print(infile)
        f = open(infile)
        fits = []
        errorr = []
        hierr = []
        loerr = []
        for line in f:
            if len(line) == 0:
                continue
            fields = line.split()
            if len(fields) == 0:
                continue
            fits.append(float(fields[0]))
            hierr.append(float(fields[2]) - float(fields[0]))
            loerr.append(float(fields[0]) - float(fields[1]))    
        f.close()

        plt.xlabel('Years since Launch')
        plt.ylabel('Line Centroid (keV)')
        #plt.plot(range(len(fits)), fits, 'o')
        plt.errorbar(0.25*np.arange(len(fits)), fits, loerr, fmt='.', label='{}'.format(det))
        break
    plt.legend()

plt.axvline(x=last_cal_year.value, color = 'r', linestyle='--')


plt.title('FPMA')
plt.show()
```
