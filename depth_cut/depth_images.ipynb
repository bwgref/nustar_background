```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import glob
import pickle
from astropy.visualization import simple_norm
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap


viridis = cm.get_cmap('viridis', 256)
newcolors = viridis(np.linspace(0, 1, 256))
newcolors[0, :] = np.array([1, 1, 1, 1])
newcmp = ListedColormap(newcolors)

viridis = cm.get_cmap('Reds', 256)
newcolors = viridis(np.linspace(0, 1, 256))
newcolors[0, :] = np.array([1, 1, 1, 1])
newcmp2 = ListedColormap(newcolors)
```

```python
for mod in 'A':
    for file in sorted(glob.glob('epoch*depth.pkl')):
        print(file, mod)
        pickle_in = open(file,"rb")
        dtab = pickle.load(pickle_in)     
        
        for det_id in [0]:
            det_key = 'det{}'.format(det_id)
            depth_im = dtab[mod][det_key]['depth_im']
            depth_im_flagged = dtab[mod][det_key]['depth_im_flaged']
            
            fig, axes = plt.subplots(nrows=3, ncols =1, figsize=[8, 16])

            norm = simple_norm(depth_im_flagged+depth_im, 'log')


            axes[0].imshow(np.transpose(depth_im_flagged+depth_im), origin = 'lower',
                       norm = norm, cmap=newcmp, extent = [0, (4096*0.04+1.6), -1000*0.04+1.6, 100*0.04+1.6], 
                      aspect = 2)
            norm = simple_norm(depth_im, 'log')

            axes[1].imshow(np.transpose(depth_im), origin = 'lower',
                       norm = norm, cmap=newcmp, extent = [0, (4096*0.04+1.6), -1000*0.04+1.6, 100*0.04+1.6], 
                      aspect = 2)
            norm = simple_norm(depth_im_flagged, 'log')
            
            axes[2].imshow(np.transpose(depth_im_flagged), origin = 'lower',
                       norm = norm, cmap=newcmp, extent = [0, (4096*0.04+1.6), -1000*0.04+1.6, 100*0.04+1.6], 
                      aspect = 2)

            

            plt.show()
            break

```
